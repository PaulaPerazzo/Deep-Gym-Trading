```python
### get stocks from s&p500
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for the Wikipedia page that lists Dow Jones companies
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# Send a GET request to fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Locate the table containing the Dow Jones components
table = soup.find('table', {'class': 'wikitable'})

# Parse the table into a pandas DataFrame
df = pd.read_html(str(table))[0]

# Print the DataFrame
print(df[['Symbol', 'Security']])
```

```python
### get stocks symbols
symbols = df['Symbol'].tolist()
symbols
```

```python
### get stocks data
stock_data = yf.download(symbols, start="2000-01-01", end="2024-01-01", group_by="ticker")
stock_data
```

```python
### get only the adjusted close prices
### get only the adj close prices
adj_close_values = []

for stock in symbols:
    adj_close_values.append(stock_data[stock]['Adj Close'])

adj_close_values = pd.DataFrame(adj_close_values).T
adj_close_values.columns = symbols
adj_close_values.drop(columns=['BRK.B', 'SW', 'SOLV', 'GEV', 'BF.B'], inplace=True)
adj_close_values
```

```python
### substitute nan values with 0
# adj_close_values.fillna(method='bfi', inplace=True)
adj_close_values.bfill(inplace=True)

## rename index name
adj_close_values.index.name = 'Ticker'
adj_close_values
```

```python
### save the data
adj_close_values.to_csv("./data/sp_stocks.csv")
```

# Metrics Analysis

```python
import pandas as pd

sp_metrics = pd.read_csv("./src/sp_500/logs/metrics_sp500.csv")
sp_metrics
```

```python
pre_pandemic_data = sp_metrics[sp_metrics["period"] == "pre-pandemic"]
pre_pandemic_data.set_index("period", inplace=True)

mean_row = pre_pandemic_data.median()
mean_row

pre_pandemic = pd.DataFrame([mean_row])
pre_pandemic["period"] = "pre-pandemic"
pre_pandemic = pre_pandemic[["period"] + [col for col in pre_pandemic.columns if col != "period"]]
pre_pandemic
```

```python
### plot the pre-pandemic data
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(data=pre_pandemic)
plt.title("Pre-Pandemic Metrics")
plt.show()
```

```python
### plot the boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=pre_pandemic)
plt.title("S&P 500 Pre-Pandemic Metrics")
plt.show()
```

```python
### pandemic period
pandemic_data = sp_metrics[sp_metrics["period"] == "pandemic"]
pandemic_data.set_index("period", inplace=True)

mean_row = pandemic_data.median()
mean_row

pandemic = pd.DataFrame([mean_row])
pandemic["period"] = "pandemic"
pandemic = pandemic[["period"] + [col for col in pandemic.columns if col != "period"]]
pandemic
```

```python
### plot the pandemic data
plt.figure(figsize=(10, 5))
sns.barplot(data=pandemic, errorbar=None)
plt.title("Pandemic Metrics")
plt.show()
```

```python
### plot the boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=pandemic_data)
plt.title("S&P 500 Pandemic Metrics")
plt.show()
```

```python
### post-pandemic period
post_pandemic_data = sp_metrics[sp_metrics["period"] == "post-pandemic"]
post_pandemic_data.set_index("period", inplace=True)

mean_row = post_pandemic_data.median()

post_pandemic = pd.DataFrame([mean_row])
post_pandemic["period"] = "post-pandemic"
post_pandemic = post_pandemic[["period"] + [col for col in post_pandemic.columns if col != "period"]]
post_pandemic
```

```python
### plot the post-pandemic data
plt.figure(figsize=(10, 5))
sns.barplot(data=post_pandemic)
plt.title("S&P 500 Post-Pandemic Metrics")
plt.show()
```

```python
### plot the boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=post_pandemic)
plt.title("S&P 500 Post-Pandemic Metrics")
plt.show()
```

```python
### all period
all_period_data = sp_metrics[sp_metrics["period"] == "all"]
all_period_data.set_index("period", inplace=True)

mean_row = all_period_data.median()
mean_row

all_period = pd.DataFrame([mean_row])
all_period["period"] = "all"
all_period = all_period[["period"] + [col for col in all_period.columns if col != "period"]]
all_period
```

```python
### plot the all period data
plt.figure(figsize=(10, 5))
sns.barplot(data=all_period)
plt.title("S&P 500 All Period Metrics")
plt.show()
```

```python
### plot the boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=all_period)
plt.title("S&P 500 All Period Metrics")
plt.show()
```

```python
### concatenate all periods data
sp_metrics = pd.concat([pre_pandemic, pandemic, post_pandemic, all_period])
sp_metrics
```

```python
### plot the distribution of the metrics
plt.figure(figsize=(10, 5))
sns.boxplot(data=sp_metrics)
plt.title("S&P 500 Metrics Distribution")
plt.show()
```

```python
### plot the distribution for each period
sp_metrics_T = sp_metrics.set_index("period")
plt.figure(figsize=(10, 5))
sns.boxplot(data=sp_metrics_T.T)
plt.title("S&P 500 Metrics Distribution")
plt.show()
```

```python
### make a subplot with boxplots for each metric
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'bov_metrics' is your DataFrame
metrics = ["arr", "avol", "mdd", "sharpe_ratio", "calmar_ratio", "sortino_ratio"]

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Loop through each metric and create a boxplot in a subplot
for i, metric in enumerate(metrics):
    df_melted = sp_metrics[["period", metric]].melt(id_vars='period', var_name='metric', value_name='value')
    sns.boxplot(x='metric', y='value', hue='period', data=df_melted, ax=axes[i])
    axes[i].set_title(f'Distribution of {metric} Across Different Periods')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
### plot a barplot with the metrics

# Set up the subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Metrics Distribution - S&P 500')

# List of metrics
metrics = ['arr', 'avol', 'mdd', 'sharpe_ratio', 'calmar_ratio', 'sortino_ratio']
colors = ['#1f77b4', '#7f7f7f', '#2ca02c', '#d62728']  # Colors for different periods

# Plot each metric in a separate subplot
for i, metric in enumerate(metrics):
    ax = axes[i // 3, i % 3] 
    # change the color of the bars according to the period
    for j, period in enumerate(sp_metrics['period'].unique()):
        sns.barplot(x='period', y=metric, data=sp_metrics[sp_metrics['period'] == period], ax=ax, color=colors[j])

        ### add the mean value as a text in the plot
        for j, period in enumerate(sp_metrics['period'].unique()):
            mean_value = sp_metrics[sp_metrics['period'] == period][metric].values[0]
            ax.text(j, mean_value, f'{mean_value:.3f}', ha='center', va='bottom')
        
    ax.set_title(metric)
    ax.set_ylabel(metric)
    ax.set_xlabel('Period')

# Adjust layout for better display
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the main title
plt.show()
```

```python
### plot a barplot with the the metrics by period (2x2 subplot)

# Set up the subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Metrics Distribution Across Different Periods')

### make the first subplot
# Plot the pre-pandemic period
ax = axes[0, 0]
sns.barplot(data=pre_pandemic, ax=ax, errorbar=None)
ax.set_title('Pre-Pandemic Period')
ax.set_ylabel('Metrics Values')
ax.set_xlabel('Metrics')
# set the y-axis limit
ax.set_ylim(0, 70)

# plot the pandemic period
ax = axes[0, 1]
sns.barplot(data=pandemic, ax=ax, errorbar=None)
ax.set_title('Pandemic Period')
ax.set_ylabel('Metrics Values')
ax.set_xlabel('Metrics')
ax.set_ylim(0, 70)

# plot the post-pandemic period
ax = axes[1, 0]
sns.barplot(data=post_pandemic, ax=ax, errorbar=None)
ax.set_title('Post-Pandemic Period')
ax.set_ylabel('Metrics Values')
ax.set_xlabel('Metrics')
ax.set_ylim(0, 70)

# plot the all period
ax = axes[1, 1]
sns.barplot(data=all_period, ax=ax, errorbar=None)
ax.set_title('All Period')
ax.set_ylabel('Metrics Values')
ax.set_xlabel('Metrics')
ax.set_ylim(0, 70)

# Adjust layout for better display
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the main title
plt.show()
```
